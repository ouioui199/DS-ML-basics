```python
from scipy import misc
import matplotlib.pyplot as plt
face = misc.face(gray=True)
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
type(face)
```

```python
face.shape
```

```python
a = face.shape[0]
b = face.shape[1]
a,b
```

```python
C = face[a/4:3*a/4,b/4:3*b/4]
```

```python
a1 = a//4
```

```python
b1 = b//4
```

```python
C = face[a1:-a1,b1:-b1]
```

```python
plt.imshow(C, cmap=plt.cm.gray)
plt.show()
```

```python
C[C>150] = 255
plt.imshow(C, cmap=plt.cm.gray)
plt.show()
```
