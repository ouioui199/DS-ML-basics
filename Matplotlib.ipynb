```python
import numpy as np
```

```python
x = np.linspace(0,2,10)
y = x**2
print(x)
```

# Pyplot

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(12,8))
```

```python
plt.plot(x,y, c='red', lw=3, ls='--', label = 'quadratique') #lw = linewidth, ls = linestyle
plt.plot(x,x**3, label = 'cubique')
plt.title('figure 1')
plt.xlabel('axe x')
plt.ylabel('axe y')
plt.legend()
plt.show()
plt.savefig('figure.png')

plt.figure()
plt.scatter(x,y)
plt.show()
```

```python
plt.scatter(x,y)
plt.show()
```

```python
plt.subplot(2,1,1)
plt.plot(x,y, c='red')
plt.subplot(2,1,2)
plt.plot(x,y, c='blue')
```

```python
plt.figure()
plt.subplot(2,2,1)
plt.plot(x, y)
plt.title('graphique 1')
plt.subplot(2,2,2)
plt.plot(x, np.sin(x), label='sinus')
plt.plot(x, np.cos(x), label='cosinus')
plt.legend()
plt.show()
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(x,y)
ax[1].plot(x, np.sin(x))
plt.show()
```

```python
dataset = {f"experience{i}": np.random.randn(100,3) for i in range (6)}
dataset
```

```python
len(dataset)
```

```python
def graphique(dataset):
    fig, ax = plt.subplots(4,1,sharex=True)
    for i in range (len(dataset)):
        ax[i].plot(dataset[f"experience{i}"])
        plt.show()
    return 0
```

```python
graphique
```

```python
graphique(dataset)
```

```python
def graphique(data):
    plt.figure(figsize=(12,30))
    n=len(data)
    for i in range (n):
        plt.subplot(n,1,i+1)
        plt.plot(data[f"experience{i}"])
        plt.title(f"experience{i}")
    plt.show()
```

```python
graphique(dataset)
```

```python
#Corrige Guillaume
def graphique(data):
    n =  len(data)
    plt.figure(figsize=(12,30))
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
    plt.show()
```

```python
graphique(dataset)
```
