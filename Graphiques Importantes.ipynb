```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
iris = load_iris()

x = iris.data
y = iris.target
names = list(iris.target_names)

print(f'x contient {x.shape[0]} exemples et {x.shape[1]} variables')
print(f'il y a {np.unique(y).size} classes')
```

# 5.Graphique de Classification avec plt.scatter

```python
plt.scatter(x[:,0], x[:,1], c=y, alpha = 0.5, s=x[:,2]*100) #alpha: transparence, s: taille
plt.xlabel('longeur sepal')
plt.ylabel('largeur sepal')
```

# 4. Graphique 3D

```python
#%matplotlib
from mpl_toolkits.mplot3d import Axes3D
```

```python
ax = plt.axes(projection='3d')
ax.scatter(x[:,0], x[:,1], x[:,2], c=y)
```

```python
f = lambda x,y: np.sin(x) + np.cos(x+y)
```

```python
X = np.linspace(0,5,100)
Y = np.linspace(0,5,100)
X,Y = np.meshgrid(X,Y)

Z = f(X,Y)
```

```python
Z.shape
```

```python
ax = plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, cmap='plasma')
```

# 3. Histogramme

```python
%matplotlib inline
plt.hist(x[:,0], bins=20)
plt.hist(x[:,1], bins=20)
```

```python
plt.hist2d(x[:,0],x[:,1], cmap='Blues')
plt.xlabel('longeur sepal')
plt.ylabel('largeur sepal')
plt.colorbar()
```

```python
from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face, cmap='gray')
plt.hist(face.ravel(),bins=255) #ravel flatten array
plt.show()
```

```python
n = x.shape[1]
plt.figure(figsize=(12,8))
for i in range (n):
    plt.subplot(n//2, n//2, i+1)
    plt.scatter(x[:,0], x[:,i], c=y)
    plt.xlabel('0')
    plt.ylabel(i)
    plt.colorbar(ticks=list(np.unique(y)))
plt.show()
```
